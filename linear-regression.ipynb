```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
train_df = pd.read_csv("../input/random-linear-regression/train.csv")
test_df = pd.read_csv("../input/random-linear-regression/test.csv")
```

```python
train_df.head()
```

```python
train_df.y.isnull().sum()
```

```python
train_df.dropna(inplace=True)
```

```python
x = np.array(train_df.x).reshape(-1,1)
y = np.array(train_df.y).reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)
```

```python
y_pred = lr.predict(x_test)
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred, color="red")
```
